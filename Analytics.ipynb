# 📊 Data Insights & Initial Findings

Based on the statistical summary (describe/info) of the dataset (N=3400 students), here are the primary insights regarding student retention, risk, and demographics.

### 1. Retention & Risk Overview
* **Retention Rate:** The overall retention rate (`RETURNED_2ND_YR`) is **78.7%**. This serves as the baseline metric for all predictive modeling.
* **Risk Distribution:**
    * The most frequent `risk_category` is **"1-Low Risk"**.
    * However, the `gateway_struggle` mean is **0.287**, indicating that **28.7%** of students struggle with foundational courses, which is a significant "at-risk" segment.
* **Attrition Cost:** The average `revenue_loss_if_not_returned` is **~$2,551** per student record (across the whole population). For the ~21.3% who leave, the loss is realized at **$12,000** per student.

### 2. Academic Performance
* **Incoming Quality:** Students are entering with a solid background, boasting an average `HIGH_SCHL_GPA` of **3.20**.
* **First-Year Momentum:**
    * The average `FIRST_TERM_EARNED_HRS` is **12.2 hours**.
    * *Insight:* Since 12 hours is often the minimum for full-time status and 15 is recommended for on-time graduation, the average student is hovering near the minimum threshold, suggesting low "Credit Momentum."
* **Course Bottlenecks:**
    * Most common Fall Core Course: **ENGL 1101** (Frequency: 878).
    * Most common Spring Core Course: **ENGL 1102** (Frequency: 897).
    * *Note:* The most frequent grade in the first core course is **"A"** (1024 students), which contrasts with the 28% gateway struggle rate. This suggests the struggle is concentrated in specific difficult courses (likely Math/Science) rather than English.

### 3. Financial Health (Critical Factor)
* **Stress Levels:** The most frequent (top) category for `financial_stress_level` is **"High Stress"**. This suggests that financial instability is a systemic issue, not an outlier.
* **Unmet Need:**
    * **Mean:** ~$72,220 (This figure appears extremely high; verify if this is an aggregate over 4 years or if the scale differs, e.g., cents vs dollars. If accurate, it implies a massive affordability gap).
    * **Extreme Need:** **~9.7%** of students (`extreme_unmet_need` mean 0.097) fall into the most critical financial category.
    * **Variability:** The standard deviation for Unmet Need is very high (~279k), indicating a massive disparity between the wealthiest and poorest students.

### 4. Demographic Profile
* **Gender:** The student body is predominantly **Female** (Top: F, Frequency: 2140/3400 or **~63%**).
* **Residency:** Heavily skewed towards locals, with **90%** (`3063/3400`) having an `IN_STATE_FLAG` of 'Y'.
* **First Generation:** Approximately **25%** (`mean: 0.249`) of the students are First-Generation college students.
* **Housing:** The majority live **Off Campus** (Top: Off Campus, Freq: 1981), which often correlates with lower engagement/retention compared to on-campus living.

### 5. Data Quality & Missing Values
* **Test Scores:** `STDNT_TEST_ENTRANCE_COMB` has **518 missing values** (Count: 2882 vs 3400). You will need an imputation strategy (e.g., fill with median) or a flag for "No Score".
* **Parental Education:** Significant missing data in `FATHER_HI_EDU_CD` (~432 missing) and `MOTHER_HI_EDU_CD` (~489 missing).
* **Second Term Attrition:** `SECOND_TERM_EARNED_HRS` has a count of 3191, meaning **209 students** (approx 6%) likely dropped out *before* completing the second term.

```python

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 10)

df = pd.read_csv('student_retention_DASHBOARD.csv')

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

demographics = [
    ('STDNT_GENDER', 'Retention by Gender', 'Blues_d'),
    ('residency_label', 'Retention by Residency', 'Greens_d'),
    ('first_gen_label', 'Retention by First Gen Status', 'Oranges_d')
]

for i, (col, title, palette) in enumerate(demographics):
    retention_rates = df.groupby(col)['RETURNED_2ND_YR'].mean().reset_index()
    
    sns.barplot(data=retention_rates, x=col, y='RETURNED_2ND_YR', ax=axes[i], palette=palette, hue=col, legend=False)
    
    axes[i].set_title(title, fontsize=14, fontweight='bold')
    axes[i].set_ylim(0, 1) 
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Retention Rate')
    
    for p in axes[i].patches:
        axes[i].annotate(f'{p.get_height():.1%}', 
                        (p.get_x() + p.get_width() / 2., p.get_height()), 
                        ha = 'center', va = 'center', 
                        xytext = (0, 9), 
                        textcoords = 'offset points',
                        fontweight='bold')

plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))

bg_summary = df.groupby('STDNT_BACKGROUND')['RETURNED_2ND_YR'].mean().reset_index()
bg_summary = bg_summary.sort_values('RETURNED_2ND_YR', ascending=True)

chart = sns.barplot(data=bg_summary, x='STDNT_BACKGROUND', y='RETURNED_2ND_YR', palette='Reds_d')

plt.title('Retention Rate by Student Background (Lowest to Highest)', fontsize=16, fontweight='bold')
plt.ylim(0, 1)
plt.ylabel('Retention Rate')
plt.xlabel('Background Category')

for p in chart.patches:
    chart.annotate(f'{p.get_height():.1%}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha = 'center', va = 'center', 
                    xytext = (0, 9), 
                    textcoords = 'offset points',
                    fontweight='bold')

plt.show()
```

```python
# Set a clean visual style for the plots
sns.set_style("whitegrid")

# --- 1. Load Data ---
# Load the main dashboard file for student-level analysis
df_main = pd.read_csv('student_retention_DASHBOARD.csv')

# Load the summary file for course-level analysis
# NOTE: Assuming the summary file column names are:
# 'Core Course Name', 'DF Rate' (D or F Rate), and 'Retention Rate'
df_gateway = pd.read_csv('summary_gateway_courses.csv')

print("--- Academic Analysis Started ---")

# ====================================================================
# A. High School GPA and Retention
# ====================================================================

# Binning GPA into quartiles or similar bins for visualization
# We can use the existing 'gpa_quartile' column if available, or create bins.
# Let's use the 'gpa_quartile' for consistency and cleaner grouping.

# Calculate retention rate by GPA quartile
gpa_retention = df_main.groupby('gpa_quartile')['RETURNED_2ND_YR'].mean().reset_index()
gpa_retention['RETURNED_2ND_YR'] = gpa_retention['RETURNED_2ND_YR'] * 100 # Convert to percentage

plt.figure(figsize=(8, 6))
chart_gpa = sns.barplot(data=gpa_retention, x='gpa_quartile', y='RETURNED_2ND_YR', palette='magma')
plt.title('Retention Rate by High School GPA Quartile', fontsize=14, fontweight='bold')
plt.ylabel('Retention Rate (%)')
plt.xlabel('HS GPA Quartile')
plt.ylim(gpa_retention['RETURNED_2ND_YR'].min() - 5, 100) # Set Y-limit dynamically

# Add data labels
for p in chart_gpa.patches:
    chart_gpa.annotate(f'{p.get_height():.1f}%',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center',
                    xytext=(0, 9),
                    textcoords='offset points',
                    fontweight='bold')

plt.show()
print("\nInsight A: The relationship between High School GPA and retention is shown.")
print(gpa_retention)
print("-" * 50)

# ====================================================================
# B. First Term Earned Hours and Retention
# ====================================================================

# Define bins for earned hours:
# 0-11 (Part-time/Struggle), 12-14 (Standard Full-Time), 15+ (On-Track/High Momentum)
bins = [0, 11, 14, df_main['FIRST_TERM_EARNED_HRS'].max() + 1]
labels = ['Struggle (<12 Hrs)', 'Standard (12-14 Hrs)', 'Momentum (15+ Hrs)']

# Create a binned column for earned hours
df_main['earned_hours_group'] = pd.cut(df_main['FIRST_TERM_EARNED_HRS'], bins=bins, labels=labels, right=False)

# Calculate retention rate by earned hours group
hours_retention = df_main.groupby('earned_hours_group', observed=True)['RETURNED_2ND_YR'].mean().reset_index()
hours_retention['RETURNED_2ND_YR'] = hours_retention['RETURNED_2ND_YR'] * 100

plt.figure(figsize=(8, 6))
chart_hours = sns.barplot(data=hours_retention, x='earned_hours_group', y='RETURNED_2ND_YR', palette='viridis')
plt.title('Retention Rate by First Term Earned Credit Hours', fontsize=14, fontweight='bold')
plt.ylabel('Retention Rate (%)')
plt.xlabel('First Term Earned Hours Group')

# Add data labels
for p in chart_hours.patches:
    chart_hours.annotate(f'{p.get_height():.1f}%',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center',
                        xytext=(0, 9),
                        textcoords='offset points',
                        fontweight='bold')

plt.show()
print("\nInsight B: Retention rates based on first-term credit momentum.")
print(hours_retention)
print("-" * 50)

# ====================================================================
# C. Gateway Course Analysis (High D/F Rates)
# ====================================================================

# Rename columns for ease of use (based on file description)
df_gateway.columns = ['Core Course Name', 'Enrollment', 'DF Rate', 'Retention Rate']

# Convert 'DF Rate' from percentage string (e.g., '47.3%') to float (47.3)
# We need to clean the percentage sign and then convert.
try:
    df_gateway['DF Rate (%)'] = df_gateway['DF Rate'].str.replace('%', '').astype(float)
except AttributeError:
    # Handle case where DF Rate might already be a float/int (if pre-cleaned)
    df_gateway['DF Rate (%)'] = df_gateway['DF Rate']

# Sort by DF Rate to find the "Killer Courses"
killer_courses = df_gateway.sort_values('DF Rate (%)', ascending=False).head(5)

plt.figure(figsize=(10, 6))
# Create a bar plot for the top 5 highest D/F rate courses
chart_killer = sns.barplot(data=killer_courses, x='DF Rate (%)', y='Core Course Name', palette='OrRd_r')
plt.title('Top 5 Gateway Courses by D/F Rate (Highest Failure)', fontsize=14, fontweight='bold')
plt.xlabel('D/F Rate (%)')
plt.ylabel('Core Course Name')

# Add data labels
for p in chart_killer.patches:
    chart_killer.annotate(f'{p.get_width():.1f}%',
                        (p.get_width(), p.get_y() + p.get_height() / 2.),
                        ha='left', va='center',
                        xytext=(5, 0),
                        textcoords='offset points',
                        fontweight='bold')
plt.xlim(0, killer_courses['DF Rate (%)'].max() + 5) # Adjust x-limit for labels
plt.show()

print("\nInsight C: The top 5 gateway courses with the highest D/F rates (potential barriers).")
print(killer_courses[['Core Course Name', 'DF Rate (%)', 'Retention Rate']])
print("-" * 50)
print("\n--- Academic Analysis Complete ---")
```

```python
# Set a clean visual style for the plots
sns.set_style("whitegrid")

# --- 1. Load Data ---
# Load the main dashboard file for student-level analysis (used for unmet need/extreme need flag)
df_main = pd.read_csv('student_retention_DASHBOARD.csv')

# Load the summary file for aggregated financial stress analysis (as requested)
# NOTE: We assume 'summary_by_financial_stress.csv' contains columns for 'financial_stress_level'
# and 'Retention Rate' (or similar retention metric).
try:
    df_financial_summary = pd.read_csv('summary_by_financial_stress.csv')
    
    # Standardize expected column names from file description (Retention Rate, Avg Unmet Need)
    # We will try to map the columns based on the description provided previously:
    # 'financial_stress_level', 'Student Count', 'Retention Rate', 'Avg Unmet Need'
    
    # --- For safety, we will calculate the summary if column names are unknown ---
    # Calculate the aggregated summary from the main file for guaranteed column consistency
    df_financial_summary_calculated = df_main.groupby('financial_stress_level', observed=True)['RETURNED_2ND_YR'].mean().reset_index()
    df_financial_summary_calculated['Retention Rate (%)'] = df_financial_summary_calculated['RETURNED_2ND_YR'] * 100
    df_financial_summary_calculated.sort_values('Retention Rate (%)', ascending=False, inplace=True)
    
except FileNotFoundError:
    print("Warning: summary_by_financial_stress.csv not found. Calculating summary from DASHBOARD file.")
    df_financial_summary_calculated = df_main.groupby('financial_stress_level', observed=True)['RETURNED_2ND_YR'].mean().reset_index()
    df_financial_summary_calculated['Retention Rate (%)'] = df_financial_summary_calculated['RETURNED_2ND_YR'] * 100
    df_financial_summary_calculated.sort_values('Retention Rate (%)', ascending=False, inplace=True)

# Use the calculated summary for plotting to ensure robustness
df_financial_plot = df_financial_summary_calculated

print("--- Financial Impact Analysis Started ---")
print("-" * 50)

# ====================================================================
# A. Distribution of Financial Stress Levels
# ====================================================================

# Calculate the count of students in each financial stress level
stress_counts = df_main['financial_stress_level'].value_counts().reset_index()
stress_counts.columns = ['financial_stress_level', 'Count']

plt.figure(figsize=(8, 6))
chart_stress_dist = sns.barplot(data=stress_counts, x='financial_stress_level', y='Count', 
                                order=df_main['financial_stress_level'].value_counts().index, 
                                palette='PuBuGn')
plt.title('Distribution of Financial Stress Levels', fontsize=14, fontweight='bold')
plt.ylabel('Student Count')
plt.xlabel('Financial Stress Level')

# Add data labels (percentage of total students)
total = stress_counts['Count'].sum()
for p in chart_stress_dist.patches:
    percentage = f'{(100 * p.get_height() / total):.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    chart_stress_dist.annotate(percentage, (x, y), ha='center', va='center', 
                            xytext=(0, 9), textcoords='offset points', fontweight='bold')

plt.show()
print("\nInsight A: Distribution of students across financial stress levels (most common state).")
print(stress_counts)
print("-" * 50)

# ====================================================================
# B. Retention Rate vs. Financial Stress Level
# (Visualizing the Inverse Relationship)
# ====================================================================

plt.figure(figsize=(8, 6))

# Plot retention rate against financial stress level (sorted by retention, descending)
chart_stress_retention = sns.barplot(data=df_financial_plot, 
                                    x='financial_stress_level', 
                                    y='Retention Rate (%)', 
                                    palette='plasma',
                                    hue='financial_stress_level',
                                    legend=False)

plt.title('Retention Rate by Financial Stress Level', fontsize=14, fontweight='bold')
plt.ylabel('Retention Rate (%)')
plt.xlabel('Financial Stress Level')
plt.ylim(df_financial_plot['Retention Rate (%)'].min() - 5, 100) # Adjust Y-limit dynamically

# Add data labels
for p in chart_stress_retention.patches:
    chart_stress_retention.annotate(f'{p.get_height():.1f}%',
                                    (p.get_x() + p.get_width() / 2., p.get_height()),
                                    ha='center', va='center',
                                    xytext=(0, 9),
                                    textcoords='offset points',
                                    fontweight='bold')

plt.show()
print("\nInsight B: Direct relationship between financial stress and retention rate.")
print(df_financial_plot[['financial_stress_level', 'Retention Rate (%)']])
print("-" * 50)

# ====================================================================
# C. Retention Rate vs. Extreme Unmet Need (High/Low Comparison)
# ====================================================================

# The 'extreme_unmet_need' column is a binary flag (0=No, 1=Yes)
extreme_need_retention = df_main.groupby('extreme_unmet_need')['RETURNED_2ND_YR'].mean().reset_index()
extreme_need_retention['Retention Rate (%)'] = extreme_need_retention['RETURNED_2ND_YR'] * 100
extreme_need_retention['Unmet Need Group'] = extreme_need_retention['extreme_unmet_need'].replace({0: 'Manageable Need', 1: 'Extreme Need'})

plt.figure(figsize=(7, 5))
chart_unmet = sns.barplot(data=extreme_need_retention, 
                        x='Unmet Need Group', 
                        y='Retention Rate (%)', 
                        palette=['#4c72b0', '#c44e52'],
                        hue='Unmet Need Group',
                        legend=False)
plt.title('Retention Rate: Manageable vs. Extreme Unmet Need', fontsize=14, fontweight='bold')
plt.ylabel('Retention Rate (%)')
plt.xlabel('')
plt.ylim(extreme_need_retention['Retention Rate (%)'].min() - 5, 100)

# Add data labels
for p in chart_unmet.patches:
    chart_unmet.annotate(f'{p.get_height():.1f}%',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center',
                        xytext=(0, 9),
                        textcoords='offset points',
                        fontweight='bold')

plt.show()
print("\nInsight C: Retention difference between students with and without extreme unmet need.")
print(extreme_need_retention[['Unmet Need Group', 'Retention Rate (%)']])
print("-" * 50)
print("\n--- Financial Impact Analysis Complete ---")
```
